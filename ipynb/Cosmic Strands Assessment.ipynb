## Step 1:Extract Word Document from Shared Location

```python
import docx2txt
# read in word file
text = docx2txt.process("comic1.docx", 'r')
```

## Step2 : Extract the contents or data from word document

```python
text  #all the words are incorrect
```

##  Before machine Learning on word file 




## Step 3: Edit the contents that are extracted from word document using Python

```python
from textblob import TextBlob
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
b = TextBlob(text)
```

```python
cor_text = b.correct()
```

```python
cor_text  ## All the words are correct now
```

## Track Changes of Word After applying  machine Learning textblob library on word file 



## Step 4: Save the contents back in same word document with Track Changes enabled

```python
## converting the paragarph in the list 

list_string = cor_text.sentences
st = ""
for i in list_string:
  print(str(i))
  st += str(i)
```

```python
print(st)
with open("comic1.docx", "w") as f:
  f.truncate()
  f.write(st)
```

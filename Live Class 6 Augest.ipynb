```python
n = int(input("Enter the number of rowa you want to print "))
for i in range(n,0,-1): ###(10,9,8,7,6,5)
    for j in range(n-i):  ### 10 - 10==0
        print(' ', end='')
        
    for j in range(2*i-1): 
        print('*',end='') 
    print()
```

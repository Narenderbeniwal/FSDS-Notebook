```python


class Solution:
    def restoreString(self, s: str, indices: List[int]) -> str:

        output = [''] * len(s) # make empty list same length as s
        index = 0
        
        for n in indices:
            output[n] = s[index]
            index += 1
        
        output = ''.join(output)
        
        return output

```
